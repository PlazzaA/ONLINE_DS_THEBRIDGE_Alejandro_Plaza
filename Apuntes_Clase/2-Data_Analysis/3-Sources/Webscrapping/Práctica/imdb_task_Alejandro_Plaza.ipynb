```python
driver.find_elements(By.CLASS_NAME, "sc-f30335b4-0.eefKuM.cli-children")
```

```python
all_li = driver.find_elements(By.CLASS_NAME, "sc-f30335b4-0.eefKuM.cli-children")
```

```python
lista = []
```

```python
for x in all_li:
    lista.append(x.find_element(By.CLASS_NAME, "ipc-title__text").text)
```

```python
for y in lista:
    y.split(".")
    print(y)
```

```python
sublista = []
for elemento in lista:
    sublista.append(elemento.split(". ", 1))
# print(sublista)
# como al sacar la lista de titulos, también me devuelve el top, divido los titulos y los tops
```

```python
top = [] 
titulo = []
for valor in sublista:
    top.append(valor[0])
    titulo.append(valor[1])
# de esas sublistas que me he creado añado a la lista top los tops y a la lista titulo los títulos
```

```python
año_duracion = []
for y in all_li:
    año_duracion.append(y.find_elements(By.CLASS_NAME, "sc-f30335b4-7.jhjEEd.cli-title-metadata-item")) 
# como año y duración tienen la misma clase y también un número que no sé que representa, añado todos los posibles valores a diferentes sublistas (año, duración, y número)
```

```python
año = []
duracion = []
for w in año_duracion:
    año.append(w[0].text)
    duracion.append(w[1].text)
# y aquí añado a la lista año el año y a la lita duracion la duracion
```

```python
puntuacion = []
for z in all_li:
    puntuacion.append(z.find_element(By.CLASS_NAME, "ipc-rating-star--rating").text) 
# aquí directamente añado los valores de cada puntuación a la lista puntuacion
```

```python
top250 = pd.DataFrame({"Top": top, "Título": titulo, "Año": año, "Duración": duracion, "Puntuación": puntuacion})
```
