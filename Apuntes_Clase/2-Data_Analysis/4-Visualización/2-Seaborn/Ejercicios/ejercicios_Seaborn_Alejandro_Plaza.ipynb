```python
# 17. Con el diagrama de violín vemos la forma de la distribución, pero no sabemos si Grass tiene muchos puntos y Fire tiene muy pocos.
# Para ver esto, usamos el swarmplot.
# Pinta el último diagrama de violines pero con un swarmplot en lugar de un violin
```

```python
pokemon3 =  pd.melt(pokemon2, id_vars=["Name", "Type 1", "Type 2"], value_vars=["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"], var_name="Stat", value_name="Valor_Total")
```

```python
plt.title(label = "Violines por tipo")
sns.violinplot(data = pokemon3, x = "Stat", y = "Valor_Total", palette = colores)
sns.swarmplot(data = pokemon3, x = "Stat", y = "Valor_Total", palette = colores, hue = "Type 1")
```

```python
# 18. Queremos ver los violines a la vez que los swarm, los podemos pintar juntos. 
# Elimina los diagramas de caja dentro de los violines con el parámetro inner = None y pon una transparencia de 0.9 en el swarmplot
# Pon un título con pyplot (como siempre)
```

```python
plt.title(label = "Violines por tipo sin caja")
sns.violinplot(data = pokemon3, x = "Stat", y = "Valor_Total", palette = colores, inner = None)
sns.swarmplot(data = pokemon3, x = "Stat", y = "Valor_Total", palette = ["Purple", "Orange", "White"], hue = "Type 1")
```
